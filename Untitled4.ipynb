```python
import torch
import torch.nn as nn
import torch.optim as optim

from torchvision import datasets, transforms
from torchvision.models import efficientnet_b0

from torch.utils.data import DataLoader

from sklearn.metrics import classification_report
```

```python
train_transform = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ColorJitter(
        brightness=0.2,
        contrast=0.2
    ),
    transforms.ToTensor(),
    transforms.Normalize(
        [0.485,0.456,0.406],
        [0.229,0.224,0.225]
    )
])

test_transform = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor(),
    transforms.Normalize(
        [0.485,0.456,0.406],
        [0.229,0.224,0.225]
    )
])
```

```python
from torch.utils.data import random_split
import copy

# --- HOCANIN ZORUNLU SINIF SIRALAMASI SÖZLÜĞÜ (0 TABANLI İNDEKS) ---
# Bilgisayarın klasör isimlerini bu sırayla eşlemesini sağlıyoruz.
hoca_class_to_idx = {
    'suv': 0,           # Hoca ID: 1 -> SUV
    'van': 1,           # Hoca ID: 2 -> VAN
    'station_wagon': 2, # Hoca ID: 3 -> STATION WAGON
    'micro': 3,         # Hoca ID: 4 -> MİCRO
    'f1': 4,            # Hoca ID: 5 -> AÇIK TEKERLEKLİ (F1 ARAÇLARI)
    'sedan': 5,         # Hoca ID: 6 -> SEDAN
    'hatchback': 6,     # Hoca ID: 7 -> HATCHBACK
    'pickup': 7         # Hoca ID: 8 -> PICK UP
}

# 1. Ana eğitim klasörünü yükle
full_train_dataset = datasets.ImageFolder('/content/dataset/train')

# ÖNEMLİ: PyTorch'un varsayılan alfabetik sıralamasını hocanın sırasıyla eziyoruz
full_train_dataset.class_to_idx = hoca_class_to_idx
full_train_dataset.classes = list(hoca_class_to_idx.keys())

# 2. Verinin %80'ini Eğitim (Train), %20'sini Doğrulama (Validation) olarak böl
train_size = int(0.8 * len(full_train_dataset))
val_size = len(full_train_dataset) - train_size

torch.manual_seed(42)
train_sub_dataset, val_sub_dataset = random_split(full_train_dataset, [train_size, val_size])

# 3. İsterler doğrultusunda transformasyonları ata
train_dataset = copy.deepcopy(train_sub_dataset)
train_dataset.dataset.transform = train_transform

val_dataset = copy.deepcopy(val_sub_dataset)
val_dataset.dataset.transform = test_transform

# 4. Asıl nihai TEST setini yükle ve onun da sıralamasını hoca standardına zorla
test_dataset = datasets.ImageFolder('/content/dataset/test', transform=test_transform)
test_dataset.class_to_idx = hoca_class_to_idx
test_dataset.classes = list(hoca_class_to_idx.keys())

print(f"Eğitim Örnek Sayısı: {len(train_dataset)}")
print(f"Doğrulama (Val) Örnek Sayısı: {len(val_dataset)}")
print(f"İzole Test Örnek Sayısı: {len(test_dataset)}")
print("Güncel Sınıf Sıralaması (Hoca Standartı):", test_dataset.classes)
```

```python
print(train_dataset.dataset.classes)
```

```python
train_loader = DataLoader(
    train_dataset,
    batch_size=32,
    shuffle=True
)

# Eğitim sürecinde overfitting kontrolü için kullanılacak loader
val_loader = DataLoader(
    val_dataset,
    batch_size=32,
    shuffle=False
)

# Hocaların test seti; eğitim döngüsünün tamamen DIŞINDA kalacak!
test_loader = DataLoader(
    test_dataset,
    batch_size=32,
    shuffle=False
)
```

```python
device = torch.device(
    "cuda" if torch.cuda.is_available() else "cpu"
)

print(device)
```

```python
model = efficientnet_b0(weights="DEFAULT")

model.classifier[1] = nn.Linear(
    1280,
    8
)

model = model.to(device)
```

```python
criterion = nn.CrossEntropyLoss()

optimizer = optim.Adam(
    model.parameters(),
    lr=0.0001
)
```

```python
train_losses = []
val_losses = []

train_accuracies = []
val_accuracies = []

epochs = 10

# Early Stopping Ayarları
best_val_loss = float('inf')
patience = 3
counter = 0

for epoch in range(epochs):

    # =========================
    # TRAIN
    # =========================
    model.train()

    running_loss = 0
    correct = 0
    total = 0

    for images, labels in train_loader:

        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()

        outputs = model(images)

        loss = criterion(outputs, labels)

        loss.backward()

        optimizer.step()

        running_loss += loss.item()

        _, predicted = outputs.max(1)

        total += labels.size(0)

        correct += predicted.eq(labels).sum().item()

    train_acc = 100 * correct / total
    train_loss = running_loss / len(train_loader)

    train_losses.append(train_loss)
    train_accuracies.append(train_acc)

    # =========================
    # VALIDATION
    # =========================
    model.eval()

    running_val_loss = 0
    correct = 0
    total = 0

    with torch.no_grad():

        for images, labels in val_loader:

            images = images.to(device)
            labels = labels.to(device)

            outputs = model(images)

            loss = criterion(outputs, labels)

            running_val_loss += loss.item()

            _, predicted = outputs.max(1)

            total += labels.size(0)

            correct += predicted.eq(labels).sum().item()

    val_acc = 100 * correct / total
    val_loss = running_val_loss / len(val_loader)

    val_losses.append(val_loss)
    val_accuracies.append(val_acc)

    # =========================
    # SONUÇLAR
    # =========================
    print(f"\nEpoch {epoch+1}/{epochs}")

    print(f"Train Loss: {train_loss:.4f}")
    print(f"Train Accuracy: {train_acc:.2f}%")

    print(f"Validation Loss: {val_loss:.4f}")
    print(f"Validation Accuracy: {val_acc:.2f}%")

    # =========================
    # BEST MODEL SAVE
    # =========================
    if val_loss < best_val_loss:

        best_val_loss = val_loss
        counter = 0

        torch.save({
    "model_state": model.state_dict(),
    "classes": full_train_dataset.classes
}, "best_model.pth")

        print("✅ En iyi model kaydedildi.")

    else:

        counter += 1

        print(f"⚠️ EarlyStopping Counter: {counter}/{patience}")

    # =========================
    # EARLY STOPPING
    # =========================
    if counter >= patience:

        print("⛔ Early stopping çalıştı!")
        break
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.plot(
    train_accuracies,
    label="Train Accuracy"
)

plt.plot(
    val_accuracies,
    label="Validation Accuracy"
)

plt.xlabel("Epoch")
plt.ylabel("Accuracy")

plt.title("Training & Validation Accuracy")

plt.legend()

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.plot(
    train_losses,
    label="Train Loss"
)

plt.plot(
    val_losses,
    label="Validation Loss"
)

plt.xlabel("Epoch")
plt.ylabel("Loss")

plt.title("Training & Validation Loss")

plt.legend()

plt.show()
```

```python
from sklearn.metrics import classification_report

# 1. Önce tüm paketi bir değişkene yüklüyoruz
checkpoint = torch.load("/content/best_model.pth", map_location=device)

# 2. Paketin içindeki gerçek model ağırlıklarını (.load_state_dict'e) veriyoruz
if isinstance(checkpoint, dict) and "model_state" in checkpoint:
    model.load_state_dict(checkpoint["model_state"])
    print("✅ Model ağırlıkları paket içerisinden başarıyla ayrıştırıldı ve yüklendi!")
else:
    model.load_state_dict(checkpoint)
    print("✅ Model ağırlıkları doğrudan yüklendi!")

model.eval()

y_true = []
y_pred = []

model.eval()

with torch.no_grad():

    for images, labels in test_loader:

        images = images.to(device)

        outputs = model(images)

        _, preds = torch.max(outputs, 1)

        y_true.extend(labels.numpy())
        y_pred.extend(preds.cpu().numpy())

print(classification_report(
    y_true,
    y_pred,
    target_names=train_dataset.dataset.classes
))
```
